```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
import matplotlib

font = {'size'   : 12}

matplotlib.rc('font', **font)
```

```python
distances_within = np.load('original/lfp/5c60add3-8d47-411c-9659-1dcfcaf28aa4/distances_within.npy')
```

```python
distances_across = np.load('original/lfp/5c60add3-8d47-411c-9659-1dcfcaf28aa4/distances_across.npy')
```

```python
distances_across_rewired = np.load('rewired/lfp/ddf4ec54-e03e-4f27-9991-2ac8dc5b036f/distances_across_circuits.npy')
```

```python
distances_across_disconnected = np.load('disconnected/lfp/c332a2fd-054f-4865-b920-eb8a90374591/distances_across_circuits.npy')
```

```python
meanDiff = np.mean(distances_within[distances_within!=0])
```

```python
mean_distances_across_rewired = np.mean(distances_across_rewired,axis=0)/meanDiff
std_distances_across_rewired = np.std(distances_across_rewired,axis=0)/meanDiff

mean_distances_across_disconnected = np.mean(distances_across_disconnected,axis=0)/meanDiff
std_distances_across_disconnected = np.std(distances_across_disconnected,axis=0)/meanDiff

mean_distances_across = np.mean(distances_across,axis=0)/meanDiff
std_distances_across = np.std(distances_across,axis=0)/meanDiff
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

ax.bar(np.arange(3),[mean_distances_across_rewired,mean_distances_across_disconnected,mean_distances_across[0]],alpha=0.5)

ax.errorbar(np.arange(3),[mean_distances_across_rewired,mean_distances_across_disconnected,mean_distances_across[0]],
            [std_distances_across_rewired,std_distances_across_disconnected,std_distances_across[0]],ls='none')
ax.set_xticklabels(['','','Rewired','','Disconnected','','Original circuit\n, half density\nelectrode'])
ax.set_ylabel('Normalized distance\n from baseline csd')
plt.tight_layout()
```
