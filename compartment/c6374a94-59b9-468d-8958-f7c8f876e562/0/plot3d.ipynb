```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
d = np.genfromtxt('electrode_csv_grid_highRes.csv',skip_header=1,delimiter=',',usecols=[1,2,3])
```

```python
x = d[:,0]
y = d[:,1]
z = d[:,2]
```

```python
x.shape
```

```python
import bluepysnap as bp
s = bp.Simulation('simulation_config.json')
c = s.circuit
```

```python
cells = c.nodes['S1nonbarrel_neurons'].get(group='hex0',properties=['x','y','z']).sample(frac=.1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(cells['x'],cells['y'],cells['z'],s=0.1)
ax.scatter(x,y,z)

# plt.xlabel('x')
# plt.xlim([3000,5000])
# plt.ylim([-2000,0])
# ax.set_zlim([-3500,-1500])
```
