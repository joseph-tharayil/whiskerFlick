```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
distances_within = np.load('../../5c60add3-8d47-411c-9659-1dcfcaf28aa4/distances_within.npy')
```

```python
distances_across_half = np.load('distances_across_evoked_scaled_0.5.npy')#[:,:-1,:-1]
distances_across_1 = np.load('distances_across_evoked_scaled_1.0.npy')#[:,:-1,:-1]
distances_across_2 = np.load('distances_across_evoked_scaled_2.0.npy')#[:,:-1,:-1]
```

```python
meanDiff = np.mean(distances_within[distances_within!=0])
```

```python
mean_distances_across = np.mean(distances_across_1,axis=0)/meanDiff
std_distances_across = np.std(distances_across_1,axis=0)/meanDiff
```

```python
mean_distances_across += np.rot90(np.fliplr(mean_distances_across))
```

```python
mean_distances_across
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.gca()
im = ax.imshow(mean_distances_across)
labels = ['iCSD','Sphere','Disk',]
ax.set_xticks(np.arange(3),labels=labels)
plt.yticks(np.arange(3),labels=labels)
fig.colorbar(im,label="Earthmover distance\nNormalized to distance between trials",location='bottom')
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(np.arange(len(mean_distances_across)),mean_distances_across)
ax.set_xticklabels(['','Half\ndensity','Quarter\ndensity','Sphere','Sphere\nhalf','Sphere\nquarter','Disk','Disk\nhalf','Disk\nquarter'])
ax.set_ylabel('Normalized distance from baseline csd')
plt.tight_layout()
```
